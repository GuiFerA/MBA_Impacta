```
import pandas as pd
%matplotlib inline
direct_link = 'http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv'
women_majors = pd.read_csv(direct_link)

#print(women_majors.info())
women_majors
```

Já troco o nome das colunas com o Rename Columns

Art and Performance, Education, English ,Foreign Languages, Health Professions, Public Administration

```
women_majors = women_majors.rename(columns={'Year':'Ano','Art and Performance':'Artes e Artes Cenicas','Education':'Educacao','English':'Ingles','Foreign Languages':'Linguas Estrangeiras','Health Professions':'Area da Saude','Public Administration':'Administracao Publica'})
```

Altero o LOC para puxar a mancha entre 50% e 100%

```
# Abaixo de 20%
#under_20 = women_majors.loc[0, women_majors.loc[0]<20]
#under_20
over_50 = women_majors.loc[0, (women_majors.loc[0]>50) & (women_majors.loc[0]<100)]
over_50
```
